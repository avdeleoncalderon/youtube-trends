```python
import pandas as pd
```

```python
# 1. Cargar los datos
df = pd.read_csv('trending_by_time.csv')
```

```python
# 2. Inspección inicial
print("--- Primeras filas ---")
print(df.head())

print("\n--- Información de tipos de datos y nulos ---")
print(df.info())
```

```python
# 3. Validación de Calidad de Datos
# Convertir trending_date a datetime (crucial para los gráficos de área)
df['trending_date'] = pd.to_datetime(df['trending_date'])

# Verificar si hay valores nulos
missing_values = df.isnull().sum()
print("\n--- Valores faltantes por columna ---")
print(missing_values)
```

```python
# 4. Análisis rápido para los requisitos de Melanie y Ashok
# ¿Cuántas categorías y países tenemos?
print(f"\nPaíses únicos: {df['region'].unique()}")
print(f"Total de categorías: {df['category_title'].nunique()}")
```

```python
# 5. Verificación de la métrica principal
total_videos = df['videos_count'].sum()
print(f"\nTotal de videos registrados en tendencias: {total_videos}")
```

```python
# Asegurarnos que la fecha sea datetime
df['trending_date'] = pd.to_datetime(df['trending_date'])

# Obtener la última fecha registrada
ultima_fecha = df['trending_date'].max()
print(f"Última fecha en el dataset: {ultima_fecha}")

# Filtrar los últimos 7 días de datos
ultima_semana = df[df['trending_date'] > (ultima_fecha - pd.Timedelta(days=7))]

# ¿Qué categorías dominaron en la última semana?
top_categorias = ultima_semana.groupby('category_title')['videos_count'].sum().sort_values(ascending=False)
print("\n--- Top Categorías Globales (Última Semana) ---")
print(top_categorias.head(5))
```

```python
# Análisis específico para United States
usa_trends = df[df['region'] == 'United States'].groupby('category_title')['videos_count'].sum().sort_values(ascending=False)

print("\n--- Top Categorías en United States (Histórico) ---")
print(usa_trends.head(5))
```
